# Feature Selection

```python
df.columns
```

```python
df = df[['id', 'title', 'overview' ,'genre']]
```

```python
df
```

```python
df.loc[:, 'tags'] = df['overview'] + df['genre']
```

```python
df
```

```python
df_new = df.drop(columns = ['overview', 'genre'])
df_new
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words = 'english')
cv
```

```python
vec = cv.fit_transform(df_new['tags'].values.astype('U')).toarray()
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs = cosine_similarity(vec)
```

```python
cs
```

```python
df_new[df_new['title'] == 'The Godfather']
```

```python
df_new[df_new['title'] == 'The Godfather'].index[0]
```

```python
dis = sorted(list(enumerate(cs[2])), reverse = True, key = lambda vec:vec[1])
for i in dis[0:5]:
    print(df_new.iloc[i[0]].title)
```

```python
def recommender(df):
    index = df_new[df_new['title'] == df].index[0]
    dis = sorted(list(enumerate(cs[index])), reverse=True, key=lambda vec: vec[1])
    print(cs[index])  # Use array indexing instead of iloc
    print(type(cs))
    for i in dis[0:5]:
        print(df_new.iloc[i[0]].title)
```

```python
recommender('Iron Man')
```
